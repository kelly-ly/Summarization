# TEXT RANK

```python
import os
from math import sqrt
import math
import pytextrank
import spacy
from operator import itemgetter

limit_phrases = 4
limit_sentences = 3
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank", last=True)

article_dir = "./my_evalue/articles"
result_dir = "./my_evalue/textrank"

for id in range(11490):
    id = str(id)
    article_name = str(id).zfill(6) + "_article.txt"
    result_name = str(id).zfill(6) + "_result.txt"
    article_path = os.path.join(article_dir, article_name)
    result_path = os.path.join(result_dir, result_name)
    
    with open(article_path, "r") as f:
        article_doc = f.read().splitlines()
        current_article = " ".join(article_doc)
    
    doc = nlp(current_article)
    
    sent_bounds = [ [s.start, s.end, set([])] for s in doc.sents]

    phrase_id = 0
    unit_vector = []

    for p in doc._.phrases:
        unit_vector.append(p.rank)

        for chunk in p.chunks:
            for sent_start, sent_end, sent_vector in sent_bounds:
                if chunk.start >= sent_start and chunk.end <= sent_end:
                    sent_vector.add(phrase_id)
                    break

        phrase_id += 1

        if phrase_id == limit_phrases:
            break
    
    sum_ranks = sum(unit_vector)
    unit_vector = [rank/sum_ranks for rank in unit_vector ]
    
    sent_rank = {}
    sent_id = 0

    for sent_start, sent_end, sent_vector in sent_bounds:
        sum_sq = 0.0
        for phrase_id in range(len(unit_vector)):

            if phrase_id not in sent_vector:
                sum_sq += unit_vector[phrase_id]**2.0

        sent_rank[sent_id] = sqrt(sum_sq)
        sent_id += 1
        
    sorted(sent_rank.items(), key=itemgetter(1)) 
    
    sent_text = {}
    sent_id = 0

    for sent in doc.sents:
        sent_text[sent_id] = sent.text
        sent_id += 1

    num_sent = 0
    summary_sent_id = []

    for sent_id, rank in sorted(sent_rank.items(), key=itemgetter(1)):
        summary_sent_id.append(sent_id)
        num_sent += 1

        if num_sent == limit_sentences:
            break

    summary = [sent_text[id] for id in sorted(summary_sent_id)]        
    result = " ".join(summary)
    
    with open(result_path,"w") as f:
        f.write(result)
        if int(id)%500==0:
            print("Write example ",result_name, ".")
```
